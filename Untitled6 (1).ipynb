```python
# WordCloud Visualization
stop = set(stopwords.words('english'))

def wordcloud_draw(dataset, color='white'):
    words = ' '.join(dataset)
    cleaned_word = ' '.join([word for word in words.split() if (word != 'news' and word != 'text')])
    wordcloud = WordCloud(stopwords=stop, background_color=color, width=2500, height=2500).generate(cleaned_word)
    plt.figure(1, figsize=(10, 7))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
# Text preprocessing functions
def remove_tags(text):
    remove = re.compile(r'')
    return re.sub(remove, '', text)

def special_char(text):
    reviews = ''
    for x in text:
        if x.isalnum():
            reviews += x
        else:
            reviews += ' '
    return reviews

def convert_lower(text):
    return text.lower()

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    return [x for x in words if x not in stop_words]

def lemmatize_word(text):
    wordnet = WordNetLemmatizer()
    return " ".join([wordnet.lemmatize(word) for word in text])
```

```python
# Applying preprocessing
dataset['Text'] = dataset['Text'].apply(remove_tags)
dataset['Text'] = dataset['Text'].apply(special_char)
dataset['Text'] = dataset['Text'].apply(convert_lower)
dataset['Text'] = dataset['Text'].apply(remove_stopwords)
dataset['Text'] = dataset['Text'].apply(lemmatize_word)
```

```python
# Feature extraction and model building
x = np.array(dataset['Text'])
y = np.array(dataset['CategoryId'])
cv = CountVectorizer(max_features=5000)
x = cv.fit_transform(dataset['Text']).toarray()
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0, shuffle=True)
```

```python
# List to store performance results
perform_list = []

# Function to run model with optimizations, including TensorFlow
def run_model_optimized(model_name, x_train, y_train, x_test, y_test, num_classes=None):
    mdl = None

    # Scikit-learn model selection with optimizations
    if model_name == 'Logistic Regression':
        mdl = LogisticRegression(solver='lbfgs', max_iter=1000)
    elif model_name == 'Random Forest':
        mdl = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0, n_jobs=-1)  # Parallel processing
    elif model_name == 'Multinomial Naive Bayes':
        mdl = MultinomialNB(alpha=1.0, fit_prior=True)
    elif model_name == 'Support Vector Classifier':
        mdl = SVC(kernel='linear', C=1, probability=True)  # Linear kernel for speed
    elif model_name == 'Decision Tree Classifier':
        mdl = DecisionTreeClassifier(criterion='entropy', random_state=0)

    # TensorFlow Neural Network Model
    elif model_name == 'TensorFlow Neural Network':
        mdl = Sequential()
        mdl.add(Dense(128, input_dim=x_train.shape[1], activation='relu'))  # Input layer
        mdl.add(Dense(64, activation='relu'))  # Hidden layer
        mdl.add(Dense(num_classes, activation='softmax'))  # Output layer for multi-class classification

        # Compile the model
        mdl.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        mdl.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)  # Fit the model

    # If TensorFlow is used, evaluate using a different method
    if model_name == 'TensorFlow Neural Network':
        loss, accuracy = mdl.evaluate(x_test, y_test, verbose=0)
        precision, recall, f1score, _ = score(y_test, mdl.predict_classes(x_test), average='micro')
    else:
        # Use OneVsRestClassifier for multi-class classification for non-TF models
        oneVsRest = OneVsRestClassifier(mdl, n_jobs=-1)  # Parallel processing for OVR
        oneVsRest.fit(x_train, y_train)

        # Prediction and scoring
        y_pred = oneVsRest.predict(x_test)

        # Accuracy and detailed metrics
        accuracy = round(accuracy_score(y_test, y_pred) * 100, 2)
        precision, recall, f1score, _ = score(y_test, y_pred, average='micro')

    # Print performance metrics
    print(f'Test Accuracy Score of {model_name}: % {accuracy}')
    print(f'Precision : {precision}')
    print(f'Recall : {recall}')
    print(f'F1-score : {f1score}')

    # Append performance to the list
    perform_list.append({
        'Model': model_name,
        'Test Accuracy': round(accuracy, 2),
        'Precision': round(precision, 2),
        'Recall': round(recall, 2),
        'F1': round(f1score, 2)
    })
```

```python
# Run all models
models = ['Logistic Regression', 'Random Forest', 'Multinomial Naive Bayes', 'Support Vector Classifier', 'Decision Tree Classifier']
for model in models:
    run_model(model)
```
